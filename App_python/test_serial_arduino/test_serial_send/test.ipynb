```python
import serial

ser = serial.Serial(port='COM5',baudrate=9600)

matrix = [[0 for x in range(4)] for y in range(7)]

def classifyData(data):
    station = int(data[0])

    if(station < 0 or station > 6):
        return -1
    
    angle = data[1]
    if(angle == 'O'):
        angle = 0
    elif(angle == 'H'):
        angle = 1
    elif(angle == 'R'):
        angle = 2
    else:
        return -1

    sign = data[2]
    value = int(data[3:])

    if(sign == '-'):
        matrix[station][angle] = -value
    else:
        matrix[station][angle] = value


while True:
    value= ser.readline()
    valueInString=str(value,'UTF-8')
    classifyData(valueInString)
    print(matrix[2][1])
```

```python
import time
import psutil

def display_usage(cpu_usage, mem_usage, bars):
    cpu_percent = cpu_usage/100
    mem_percent = mem_usage/100

    cpu_bars = '█' * int(cpu_percent * bars) + '-' * (bars - int(cpu_percent * bars))
    mem_bars = '█' * int(mem_percent * bars) + '-' * (bars - int(mem_percent * bars))

    print(f'\r CPU Usage: {cpu_bars} {cpu_usage:.2f}%  ', end='')
    print(f'Memory Usage: {mem_bars} {mem_usage:.2f}%  ', end='\r')

while True:
    bars = 20
    while True:
        cpu_usage = psutil.cpu_percent()
        mem_usage = psutil.virtual_memory().percent

        display_usage(cpu_usage, mem_usage, bars)
        time.sleep(0.5)
```
